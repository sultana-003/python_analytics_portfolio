## Data Analysis and Visualization

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize=True)
```

```python
cancelled_percentage
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','Cancelled'],df['is_canceled'].value_counts(), edgecolor = 'k', width =0.7)
plt.show()
```

#### Reservation Status Count

##### (Confirmed vs Canceled bookings)

This analysis shows how many reservations were successfully completed compared to those that were canceled.
The results highlight that while a majority of guests do complete their bookings, a significant portion cancel before arrival. This indicates that cancellations are an important factor affecting hotel occupancy and revenue. Understanding cancellation behavior helps hotels plan better pricing, staffing, and overbooking strategies.

```python
plt.figure(figsize = (8,4))
sns.countplot(x = 'hotel', hue = 'is_canceled', data= df)
plt.title('Reservation status of different hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

### Reservation Status of Different Hotels

#### (City Hotel vs Resort Hotel)

When comparing different hotel types, we observe noticeable differences in cancellation behavior. One hotel type experiences a higher cancellation rate than the other.
This suggests that guest expectations, booking purpose (business vs leisure), or pricing sensitivity may vary by hotel type. Such differences indicate that hotels should not apply a single strategy across all properties.

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(10,6))
plt.title('Average daily rate in hotels')
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'Resort Hotel')
plt.show()
```

### Average Daily Rate (ADR) in Hotels

The Average Daily Rate represents the average price paid per room per night.
From the analysis, we can see that one hotel type generally charges higher room rates than the other. This reflects differences in location, services offered, and target customers. Higher ADR does not always mean higher profit, especially if it leads to more cancellations.

```python
df['month'] = df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='month', hue ='is_canceled', data = df)
plt.title('reservation status per month')
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

### Reservation Status per Month

This visualization shows how bookings and cancellations change throughout the year.
Certain months experience higher cancellation rates, while others show more stable bookings. This pattern indicates seasonality in customer behavior. Guests are more likely to cancel during peak or uncertain travel seasons, while bookings are more reliable during off-peak months.

```python
plt.figure(figsize=(10,6))
plt.title('ADR per month')
sns.barplot(x ='month', y ='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

### Average Daily Rate (ADR) per Month

The monthly ADR trend shows how room prices fluctuate over the year.
Prices tend to rise during high-demand seasons and drop during low-demand periods. This confirms that hotels adjust pricing based on expected demand. However, higher prices in peak months may also contribute to higher cancellation rates.

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(6,6))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

### Top 10 Countries with Reservation Cancellations

This analysis identifies the countries where the highest number of cancellations originate.
Guests from certain countries cancel more frequently than others, possibly due to visa issues, travel distance, economic factors, or flexible booking habits. This insight helps hotels understand international customer behavior and manage risk more effectively.

### Final Report Summary

#### Based on the analysis of the hotel booking dataset, several key insights emerge:

1. Cancellations form a significant portion of total reservations and cannot be ignored.
2. Different hotel types experience different booking and cancellation patterns.
3. Room prices vary by hotel type and season, influencing guest behavior.
4. Booking and cancellation trends are strongly seasonal.
5. International guests from certain countries are more likely to cancel.

#### These patterns directly impact hotel revenue, occupancy planning, and pricing strategies.

### Recommendations

##### 1. Improve Cancellation Policies
Introduce partial prepayments or stricter cancellation deadlines during high-risk periods to reduce last-minute cancellations.
##### 2. Hotel-Specific Strategies
Develop separate pricing and marketing strategies for different hotel types instead of using a single approach.
##### 3. Dynamic Pricing Optimization
Avoid excessively high prices during peak seasons, as they may increase cancellations. Balance price and booking stability.
##### 4. Seasonal Planning
Increase marketing efforts and flexible offers during low-demand months to improve occupancy.
##### 5. Country-Based Risk Management
For countries with high cancellation rates, consider requiring deposits or offering non-refundable discounted rates.
##### 6. Better Demand Forecasting
Use historical cancellation and ADR trends to improve forecasting and resource allocation.
